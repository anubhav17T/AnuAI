### Import Files

```python
# Import Files
import numpy as np
import pandas as pd
import random
import os
import time
import sys
import matplotlib
import matplotlib.pyplot as plt
from collections import defaultdict
from sklearn.svm import SVR
import pickle
from sklearn.metrics import mean_squared_error
from sklearn.neural_network import MLPRegressor
from math import sqrt


plt.grid(True)
%matplotlib inline
```

### Reading the dataset

```python
datafile = "data.txt"
```

```python
f = open(datafile, "r")
data = f.read()
```

```python
rows = data.split("\n")
```

```python
# Deleting last element from list, as it is an empty string
del(rows[31])
```

```python
# Length of rows
len(rows)
```

```python
rows[0]
```

```python
# Creating an empty dictionary, to store the data to be used from file
day_dict = defaultdict(list)

for row in rows:
    if len(row)>1:
        day, year, month, actualtime, crsetime, airtime, ardelay, depdelay, distance = row.split("\t")

        day_dict[day].append((actualtime, crsetime, airtime, ardelay, depdelay, distance))
```

### Checking the dictionary for data arrangement

```python
#actualtime, crsetime, airtime, ardelay, depdelay, distance
actualtime_vals = []
crsetime_vals = []
airtime_vals = []
ardelay_vals = []
depdelay_vals = []
distance_vals = []


for key in sorted(day_dict):
    values = day_dict[key]
    actualtime_vals.append(float(values[0][0]))
    crsetime_vals.append(float(values[0][1]))
    airtime_vals.append(float(values[0][2]))
    ardelay_vals.append(float(values[0][3]))
    depdelay_vals.append(float(values[0][4]))
    distance_vals.append(float(values[0][5]))
```

```python
days = []
for i in range(1,32):
    days.append(i)
print(days)
```

### Applying Neural Networks to predict the Arrival and Departure Delays, based on other factors

#### These delays can help us to prepare beforehand to avoid these, and find a pattern in them, along with the financial savings a company can have, by distributing the resources efficiently

```python
len(actualtime_vals)
```

```python
# Preparing final data for neural network
ar_input_data_total = []
ar_test_input_data_total = []

dep_input_data_total = []
dep_test_input_data_total = []

ar_input_data = []
ar_outputs = []
dep_input_data = []
dep_outputs = []

ar_test_input_data = []
ar_test_outputs = []

dep_test_input_data = []
dep_test_outputs = []



# For Arrival Delay

# As we have divided the data in 31
for i in range(28):
    ar_input_data = []
    ar_input_data.append(days[i])
    ar_input_data.append(actualtime_vals[i])
    ar_input_data.append(crsetime_vals[i])
    ar_input_data.append(airtime_vals[i])
    ar_input_data.append(distance_vals[i])
    
    ar_input_data_total.append(ar_input_data)
    
    
    ar_outputs.append(ardelay_vals[i])
    
    
for i in range(28, 31):
    ar_test_input_data = []
    ar_test_input_data.append(days[i])
    ar_test_input_data.append(actualtime_vals[i])
    ar_test_input_data.append(crsetime_vals[i])
    ar_test_input_data.append(airtime_vals[i])
    ar_test_input_data.append(distance_vals[i])
    ar_test_outputs.append(ardelay_vals[i])
    
    ar_test_input_data_total.append(ar_test_input_data)
    
    
# For Departure Delay
for i in range(28):
    dep_input_data = []
    dep_input_data.append(days[i])
    dep_input_data.append(actualtime_vals[i])
    dep_input_data.append(crsetime_vals[i])
    dep_input_data.append(airtime_vals[i])
    dep_input_data.append(distance_vals[i])
    dep_outputs.append(depdelay_vals[i])
    
    dep_input_data_total.append(dep_input_data)
    
for i in range(28, 31):
    dep_test_input_data = []
    dep_test_input_data.append(days[i])
    dep_test_input_data.append(actualtime_vals[i])
    dep_test_input_data.append(crsetime_vals[i])
    dep_test_input_data.append(airtime_vals[i])
    dep_test_input_data.append(distance_vals[i])
    dep_test_outputs.append(ardelay_vals[i])
    
    dep_test_input_data_total.append(dep_test_input_data)
```

### 1. For arrival

```python
# Neural Network Regression using Multi Layer Perceptron
mlp_ann = MLPRegressor(max_iter=10000)
ar_ann = mlp_ann.fit(ar_input_data_total, ar_outputs)
ar_predict_ann = ar_ann.predict(ar_test_input_data_total)
```

```python
mse2 = mean_squared_error(ar_predict_ann, ar_test_outputs)
mse2_scaled = sqrt(mse2)
print(mse2_scaled)
```

### 2. For departure

```python
# Neural Network Regression using Multi Layer Perceptron
mlp_ann = MLPRegressor(max_iter=10000)
dep_ann = mlp_ann.fit(dep_input_data_total, dep_outputs)
dep_predict_ann = dep_ann.predict(dep_test_input_data_total)
```

```python
mse2 = mean_squared_error(dep_predict_ann, dep_test_outputs)
mse2_scaled = sqrt(mse2)
print(mse2_scaled)
```

### Using SVR for the prediction

### 1. For arrival delays

```python
# Defining 1 degree SVM Regression
clf_1 = SVR(C=0.2, epsilon=0.2, kernel='poly', degree=1)

# Fitting SVR
clf_1.fit(ar_input_data_total, ar_outputs)

# Saving SVR Model
check = pickle.dumps(clf_1)

# Predicting the list
clf_1_results = clf_1.predict(ar_test_input_data_total).tolist()


# Root Mean Square Error
mse = mean_squared_error(clf_1_results, ar_test_outputs)
mse_scaled = sqrt(mse)
print(mse_scaled)
```

### 2. For Departure Delays

```python
# Defining 1 degree SVM Regression
clf_1 = SVR(C=0.2, epsilon=0.2, kernel='poly', degree=1)

# Fitting SVR
clf_1.fit(dep_input_data_total, dep_outputs)

# Saving SVR Model
check = pickle.dumps(clf_1)

# Predicting the list
clf_1_results = clf_1.predict(dep_test_input_data_total).tolist()


# Root Mean Square Error
mse = mean_squared_error(clf_1_results, dep_test_outputs)
mse_scaled = sqrt(mse)
print(mse_scaled)
```
